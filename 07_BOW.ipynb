### Bag of words (BOW) 


converting the processed text into vectors 



-> one of the most used technique to convert the text --> vectors 

```python
import pandas as pd
```

```python

#reading the csv file
dataframe=pd.read_csv('07_spam.csv',encoding='latin-1')

#droping the three unamed columns 
dataframe = dataframe.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)

# displaying top 5 element of csv file
dataframe.head()
```

#### Text Preprocessing on message of dataset


```python
# importing re (regular expression) for removing special characters

import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer,PorterStemmer

lemmatizer=WordNetLemmatizer()
poterstemmer=PorterStemmer()
```

```python
# using porterstemmer(PorterStemmer)

corpus=[]

for i in range(0,len(dataframe)):
    review=re.sub('[^a-zA-Z]'," ",dataframe['message'][i])
    review=review.lower()
    review=review.split()

    review=[ poterstemmer.stem(word) for word in review if  not word in stopwords.words('english')]

    review=' '.join(review)

    corpus.append(review)

corpus
```

```python

# using lemmatizer (WordNetLemmatizer)
corpus2=[]

for i in range(0,len(dataframe)):
    review=re.sub('[^a-zA-Z]'," ",dataframe['message'][i])
    review=review.lower()
    review=review.split()

    review=[ lemmatizer.lemmatize(word,pos='v') for word in review if  not word in stopwords.words('english')]

    review=' '.join(review)

    corpus2.append(review)

corpus2
```

### Creating bag of words (BOW) model


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=200,binary=True)

x=cv.fit_transform(corpus2).toarray()
```

```python

# text ---> vector
print("text ---> vector\n")
print(x)

print("\nshape of vector\n")
print(x.shape)
```
